```python
import pandas as pd
import gzip
import io
```

```python
d = pd.read_csv('./in/in_new.snp', sep='\t')
f=gzip.open('./out_wbs/D_num.site.gz','r')
site=str(f.read())[2:-3]
df2 = pd.DataFrame([[x.split('\\t')[0], int(x.split('\\t')[1])] for x in site.split('\\n')], columns=['chr', 'pos'])
rs = pd.merge(df2,d, on=['chr', 'pos'], how='left').id_.to_list()
f=gzip.open('./out_wbs/D_num.TriangleV.gz','r')
txt = str(f.read())
data = [x.split('\\t')for x in txt.split('\\n')[1:-1]]

new = []
new2 = []
for idx, x in enumerate(data):
  add = []
  for j in new:
    add.append(j[idx-1])
  new2.append(add + [1] + x)
  add.extend(x)
  new.append(add)
else:
  add = []
  for j in new:
    add.append(j[idx])
  print(add)
  new2.append(add + [1])

df = pd.DataFrame(new2, columns=rs, index=rs)
```
